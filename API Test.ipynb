```python
types_df = pd.read_csv("Stores_csv.csv")
types_df
```

```python
# set up additional columns to hold information
types_df['address'] = ""
types_df.head()
```

```python
# find the closest restaurant of each type to coordinates

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
```

```python
params = {
    "location": "33.753746, -84.386330",  # ATL coords,
    "rankby": "distance",
    "types": 'store',
    "key": gkey,
}
# use iterrows to iterate through pandas dataframe
for index, row in types_df.iterrows():

    # get restaurant type from df
    store_type = row['Store Type']

    # add keyword to params dict
    params['keyword'] = 'store'

    # assemble url and make API request
    print(f"Retrieving Results for Index {index}:{store_type}")
    response = requests.get(base_url, params=params).json()
    
    # extract results
    results = response['results']
    
    try:
        print(f"The closest {store_type} is located at {results[0]['vicinity']}.")
        
        types_df.loc[index, 'address'] = results[0]['vicinity']
         
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
```
